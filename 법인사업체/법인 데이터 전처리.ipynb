## 데이터 개요
- 총 2,450 개의 동대문구 소재 회사를 [국민연금 가입 사업장 내역 데이터](https://www.data.go.kr/data/15083277/fileData.do)로부터 추려냄
- 그리고 그 회사들의 사명과 도로명 주소(도로명 까지만 있고 상세 주소는 기재되어있지 않음)를 가지고 카카오 지도 API 를 활용해 1,194 개 회사의 위도, 경도 정보를 획득함

## 고민되는 점
- 정확한 위도, 경도가 존재하지 않은 1,259 건의 법인 위/경도의 결측치를 어떻게 해야할지 고민이 됨.
